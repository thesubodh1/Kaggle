```python
# lets import the tools that we might need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
len(data)
```

```python
# lets split the data into feature and label matrix (X and Y)
X = data.drop("exam_score",axis=1)
Y = data["exam_score"]
```

```python
# There are not any missing data but there are some objects which is required to be converted into numeric value 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_features = ["gender","course","internet_access","sleep_quality","study_method","facility_rating","exam_difficulty"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,cat_features)],remainder="passthrough")
transformed_x = transformer.fit_transform(X)
transformed_x
```

```python
# lets split the data into train and test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(transformed_x,Y)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=10)
reg.fit(X_train,Y_train)
reg.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(X_train,Y_train)
lasso.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=10)
reg.fit(X_train,Y_train)
reg_preds = reg.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(Y_test,reg_preds)
```

```python
def tune_n_score():
    model = Ridge()
    alpha = np.logspace(-3,3,50)

    for value in alpha:
        model.set_params(alpha=value)
        model.fit(X_train,Y_train)
        preds = model.predict(X_test)
        print(root_mean_squared_error(Y_test,preds))

tune_n_score()
```

```python
# ok so we got best in our baseline
model = Ridge()
model.fit(X_train,Y_train)
preds = model.predict(X_test)
root_mean_squared_error(Y_test,preds)
```
