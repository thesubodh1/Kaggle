```python
# lets fill the numeric column with median and convert object into cateogry
# we are going to define a function
def preprocess_data(df):
    for labels,content in df.items():
        # fill all the missing numeric values with median
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[labels] = content.fillna(content.median())

        # convert object into cat and fill missing
        if not pd.api.types.is_numeric_dtype(content):
            df[labels] = pd.Categorical(content).codes + 1
    return df
            
```

```python
# lets preprocess the data
df = preprocess_data(data)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.Name
```

# we are going to train two models
1. Random Forest Classifier
2. Linear Regression
   

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,recall_score,f1_score
from sklearn.model_selection import cross_val_score,train_test_split
```

```python
# 1. Random Forest CLassifier
# split the data into feature and label i.e X and Y
X = df.drop("Survived",axis=1)
Y = df["Survived"]

# split the model into train and test splits
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)

rf_model = RandomForestClassifier()

rf_model.fit(X_train,Y_train)
```

```python
rf_preds =rf_model.predict(X_test)
```

```python
rf_model.score(X_test,Y_test)
```

```python
def scoring_models(y_preds):
    scores = {"Accuracy":accuracy_score(Y_test,y_preds),
              "Recall score":recall_score(Y_test,y_preds),
             "f1 score":f1_score(Y_test,y_preds)}
    return scores
```

```python
rf_scores = scoring_models(rf_preds)
rf_scores
```

```python
# 2 logistic regressor
lr_model = LogisticRegression(max_iter=2000)

lr_model.fit(X_train,Y_train)

lr_model.score(X_test,Y_test)
```

```python
lr_preds =lr_model.predict(X_test)
```

```python
lg_scores = scoring_models(lr_preds)
lg_scores
```

```python
# Hyperparameter Tuning
rf_param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"],
    "bootstrap": [True]
}

lr_param_grid = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "penalty": ["l1", "l2"],
    "solver": ["liblinear", "saga"],
    "class_weight": [None, "balanced"],
    "max_iter": [1000]
}
```

```python
# hyperparameter tuning using randomized search cv for random forest classifier
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
rf_rs = RandomizedSearchCV(estimator=RandomForestClassifier(),param_distributions=rf_param_grid,n_iter=30,verbose=True)

rf_rs.fit(X_train,Y_train)
rf_rs_preds = rf_rs.predict(X_test)
rf_rs.score(X_test,Y_test)
```

```python
rf_rs_score = scoring_models(rf_rs_preds)
rf_rs_score
```

```python
rf_rs.best_params_
```
